# Sentiment analysis.

check Polarity, and subjectivity using Textblob.

Plot sentiment analysis for overall speech.

Plot sentiment analysis for a full speech in regular intervals.

```python
import pandas as pd

corpus = pd.read_pickle('corpus.pkl')
```

```python
from textblob import TextBlob

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

corpus['polarity'] = corpus['transcript'].apply(pol)
corpus['subjectivity'] = corpus['transcript'].apply(sub)
corpus['full_name'] = ['Deepak Ravishankar Ramkumar', 'Karthik Kottiswaran',
       'Karthikeyan Narayanaswami', 'Krishna Kumar', 'Raj Kumar',
       'Raj Kumar S', 'Ram', 'Ram Balaji Subbaiyan', 'Riyas Khan',
       'Selva Vignesh']
corpus
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [25,20]
plt.title('sentiment analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)
for comedian in corpus.index:
    x = corpus.polarity.loc[comedian]
    y = corpus.subjectivity.loc[comedian]
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001, corpus['full_name'][comedian], fontsize=10)
    plt.xlim(-.01, .12)

plt.show()
```
